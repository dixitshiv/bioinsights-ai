```python
# Cell 1: Setup and load search engine
import sys
import os
sys.path.append('../src')

# Change to the correct directory
os.chdir('..')  # Go up one level to the main project directory

from similarity_search import ProteinSimilaritySearch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import time

print("🔍 BioInsights AI - Day 5: Search Engine Analysis")
print("="*55)

# Check if the file exists
embeddings_file = "models/embeddings_for_faiss.pkl"
if os.path.exists(embeddings_file):
    print(f"✅ Found embeddings file: {embeddings_file}")
else:
    print(f"❌ Embeddings file not found: {embeddings_file}")
    print("Current directory:", os.getcwd())
    print("Files in models/:", os.listdir("models/") if os.path.exists("models/") else "models/ not found")

# Initialize search engine
search_engine = ProteinSimilaritySearch()
stats = search_engine.get_database_stats()

print(f"📊 Database loaded:")
for key, value in stats.items():
    print(f"  • {key}: {value}")
```

```python
# Cell 2: Performance benchmarking
print("⚡ Performance Benchmarking:")
print("="*30)

# Test search performance with different proteins
test_proteins = search_engine.protein_ids[:5]  # Test first 5 proteins
performance_results = []

for protein_id in test_proteins:
    print(f"\n🧪 Testing {protein_id}:")
    
    # Time search
    start_time = time.time()
    results = search_engine.search_similar(query_protein_id=protein_id, top_k=5)
    search_time = time.time() - start_time
    
    print(f"  • Search time: {search_time:.4f}s")
    print(f"  • Results found: {len(results)}")
    print(f"  • Top similarity: {results[0]['similarity']:.3f}")
    
    performance_results.append({
        'protein_id': protein_id,
        'search_time': search_time,
        'results_found': len(results),
        'top_similarity': results[0]['similarity'] if results else 0
    })

# Visualize performance
perf_df = pd.DataFrame(performance_results)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(range(len(perf_df)), perf_df['search_time'], color='skyblue')
plt.title('⚡ Search Performance')
plt.ylabel('Time (seconds)')
plt.xlabel('Test Protein')
plt.xticks(range(len(perf_df)), [p.split('_')[0] for p in perf_df['protein_id']], rotation=45)

plt.subplot(1, 2, 2)
plt.bar(range(len(perf_df)), perf_df['top_similarity'], color='lightcoral')
plt.title('🎯 Search Accuracy (Self-Similarity)')
plt.ylabel('Similarity Score')
plt.xlabel('Test Protein')
plt.xticks(range(len(perf_df)), [p.split('_')[0] for p in perf_df['protein_id']], rotation=45)

plt.tight_layout()
plt.show()

print(f"\n📈 Performance Summary:")
print(f"  • Average search time: {perf_df['search_time'].mean():.4f}s")
print(f"  • Fastest search: {perf_df['search_time'].min():.4f}s")
print(f"  • Average self-similarity: {perf_df['top_similarity'].mean():.3f}")
```

```python
# Cell 3: Drug similarity matrix
print("💊 Drug-to-Drug Similarity Analysis:")
print("="*40)

# Get unique drugs
unique_drugs = list(set([meta['drug_name'] for meta in search_engine.metadata]))
print(f"Analyzing {len(unique_drugs)} drugs: {unique_drugs}")

# Create similarity matrix
similarity_matrix = np.zeros((len(unique_drugs), len(unique_drugs)))

for i, drug1 in enumerate(unique_drugs):
    # Find a representative protein for this drug (heavy chain preferred)
    drug1_proteins = [pid for pid in search_engine.protein_ids 
                     if search_engine.protein_lookup[pid]['drug_name'] == drug1]
    
    # Prefer heavy chain
    rep1 = None
    for pid in drug1_proteins:
        if 'heavy' in pid:
            rep1 = pid
            break
    if rep1 is None:
        rep1 = drug1_proteins[0]  # Use first available
    
    # Search for similar proteins
    results = search_engine.search_similar(query_protein_id=rep1, top_k=20)
    
    for result in results:
        drug2 = result['drug_name']
        if drug2 in unique_drugs:
            j = unique_drugs.index(drug2)
            similarity_matrix[i, j] = result['similarity']

# Create interactive heatmap
fig = px.imshow(similarity_matrix, 
                x=unique_drugs, y=unique_drugs,
                color_continuous_scale='RdYlBu_r',
                title="💊 Drug-to-Drug Similarity Matrix")
fig.show()

# Show most similar pairs
print(f"\n🏆 Most Similar Drug Pairs:")
for i in range(len(unique_drugs)):
    for j in range(i+1, len(unique_drugs)):
        sim = similarity_matrix[i, j]
        if sim > 0.8:  # High similarity threshold
            print(f"  • {unique_drugs[i]} ↔ {unique_drugs[j]}: {sim:.3f}")
```

```python
# Cell 4: Search validation
print("✅ Search Engine Validation:")
print("="*30)

validation_checks = {
    "FAISS index working": True,
    "Search returns results": True,
    "Self-similarity = 1.0": True,
    "Filters work correctly": True,
    "Performance under 1s": True
}

# Test each validation check
try:
    # Test basic search
    test_results = search_engine.search_similar(
        query_protein_id=search_engine.protein_ids[0], 
        top_k=3
    )
    validation_checks["Search returns results"] = len(test_results) > 0
    
    # Test self-similarity
    self_sim = test_results[0]['similarity'] if test_results else 0
    validation_checks["Self-similarity = 1.0"] = abs(self_sim - 1.0) < 0.001
    
    # Test filters
    heavy_results = search_engine.search_similar(
        query_protein_id=search_engine.protein_ids[0],
        top_k=3,
        filter_chain_type="heavy"
    )
    all_heavy = all(r['chain_type'] == 'heavy' for r in heavy_results)
    validation_checks["Filters work correctly"] = all_heavy
    
    # Test performance
    start = time.time()
    perf_results = search_engine.search_similar(
        query_protein_id=search_engine.protein_ids[0], 
        top_k=5
    )
    search_time = time.time() - start
    validation_checks["Performance under 1s"] = search_time < 1.0
    
except Exception as e:
    print(f"❌ Error during validation: {e}")

# Display results
print("Validation Results:")
for check, passed in validation_checks.items():
    status = "✅" if passed else "❌"
    print(f"  {status} {check}")

all_passed = all(validation_checks.values())
print(f"\n🎯 Overall Status: {'✅ Ready for Web Interface!' if all_passed else '❌ Needs fixes'}")

if all_passed:
    print(f"\n🚀 Your search engine is ready for Days 8-11:")
    print(f"  • Lightning-fast similarity search: ✅")
    print(f"  • Professional performance (<1s): ✅") 
    print(f"  • Accurate results (self-sim = 1.0): ✅")
    print(f"  • Working filters: ✅")
    print(f"  • Ready for Streamlit: ✅")
    
# Quick stats summary
print(f"\n📊 Final Stats:")
print(f"  • Database size: {len(search_engine.protein_ids)} proteins")
print(f"  • Unique drugs: {len(set([m['drug_name'] for m in search_engine.metadata]))}")
print(f"  • Search speed: ~{perf_df['search_time'].mean():.4f}s average")
print(f"  • Memory usage: {search_engine.embeddings.nbytes / 1024**2:.1f} MB")
```

```python
# Complete Day 5-7 Analysis - Fixed Paths
import sys
import os

print("🔍 BioInsights AI - Complete Day 5-7 Analysis")
print("="*60)

# Debug: Show current location
print(f"Current directory: {os.getcwd()}")
print(f"Directory contents: {os.listdir('.')}")

# Navigate to the correct project directory
if 'notebooks' in os.getcwd():
    os.chdir('..')
    print(f"Changed to: {os.getcwd()}")
elif 'bioinsights-ai' not in os.getcwd():
    # Try to find the project directory
    if os.path.exists('/Users/shivamdixit/Desktop/bioinsights-ai'):
        os.chdir('/Users/shivamdixit/Desktop/bioinsights-ai')
        print(f"Found and changed to: {os.getcwd()}")
    else:
        print("❌ Cannot find bioinsights-ai directory")
        print("Please navigate to your project directory manually")

# Verify we're in the right place
print(f"\nFinal working directory: {os.getcwd()}")
print("Contents:", os.listdir('.'))

# Check for required directories and files
required_paths = ['models', 'src', 'data']
for path in required_paths:
    if os.path.exists(path):
        print(f"✅ Found directory: {path}")
        if path == 'models':
            print(f"   Models contents: {os.listdir(path)}")
        elif path == 'src':
            print(f"   Src contents: {os.listdir(path)}")
    else:
        print(f"❌ Missing directory: {path}")

# Add src to Python path
src_path = os.path.join(os.getcwd(), 'src')
if src_path not in sys.path:
    sys.path.insert(0, src_path)
print(f"Added to Python path: {src_path}")

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import pickle
import faiss

print("✅ All libraries imported successfully")

# Check if embedding file exists
embedding_file = "models/embeddings_for_faiss.pkl"
if os.path.exists(embedding_file):
    print(f"✅ Found embeddings file: {embedding_file}")
else:
    print(f"❌ Embeddings file not found: {embedding_file}")
    
    # Check for alternative files
    if os.path.exists("models"):
        model_files = os.listdir("models")
        print(f"Available model files: {model_files}")
        
        # Look for any pickle files
        pkl_files = [f for f in model_files if f.endswith('.pkl')]
        if pkl_files:
            print(f"Found pickle files: {pkl_files}")
            # Try using the first available embeddings file
            for pkl_file in pkl_files:
                if 'embedding' in pkl_file.lower():
                    embedding_file = f"models/{pkl_file}"
                    print(f"🔄 Using alternative file: {embedding_file}")
                    break

# If we still don't have embeddings, we need to create them
if not os.path.exists(embedding_file):
    print("❌ No embeddings file found. Need to run embedding generation first.")
    print("\n🔧 SOLUTION: Run this in terminal:")
    print("cd /Users/shivamdixit/Desktop/bioinsights-ai")
    print("python src/embedding_generator.py")
    print("\nThen come back and run this notebook again.")
    
    # Stop here if no embeddings
    raise FileNotFoundError("Embeddings file not found. Please run embedding generation first.")

# Create inline search engine if the file exists
class ProteinSimilaritySearch:
    def __init__(self, embeddings_file=None):
        if embeddings_file is None:
            embeddings_file = embedding_file
            
        print(f"🔍 Loading embeddings from: {embeddings_file}")
        
        with open(embeddings_file, "rb") as f:
            data = pickle.load(f)
        
        self.embeddings = data['embeddings']
        self.protein_ids = data['protein_ids']
        self.metadata = data['metadata']
        self.protein_lookup = {pid: meta for pid, meta in zip(self.protein_ids, self.metadata)}
        
        print(f"✅ Loaded {len(self.protein_ids)} proteins")
        print(f"✅ Embedding dimension: {self.embeddings.shape[1]}")
        
        # Build FAISS index
        embeddings_normalized = self.embeddings / np.linalg.norm(self.embeddings, axis=1, keepdims=True)
        dimension = embeddings_normalized.shape[1]
        self.index = faiss.IndexFlatIP(dimension)
        self.index.add(embeddings_normalized.astype('float32'))
        
        print(f"✅ FAISS index built with {self.index.ntotal} vectors")
        
    def search_similar(self, query_protein_id=None, query_embedding=None, top_k=5, 
                      filter_chain_type=None, **kwargs):
        start_time = time.time()
        
        if query_protein_id:
            if query_protein_id not in self.protein_ids:
                raise ValueError(f"Protein {query_protein_id} not found")
            idx = self.protein_ids.index(query_protein_id)
            query_embedding = self.embeddings[idx]
        
        if query_embedding is None:
            raise ValueError("Must provide query_protein_id or query_embedding")
            
        query_embedding_norm = query_embedding / np.linalg.norm(query_embedding)
        query_embedding_norm = query_embedding_norm.astype('float32').reshape(1, -1)
        
        similarities, indices = self.index.search(query_embedding_norm, min(top_k * 3, len(self.protein_ids)))
        
        results = []
        for sim, idx in zip(similarities[0], indices[0]):
            if idx == -1:
                continue
                
            protein_id = self.protein_ids[idx]
            metadata = self.protein_lookup[protein_id]
            
            if filter_chain_type and metadata['chain_type'] != filter_chain_type:
                continue
            
            results.append({
                'protein_id': protein_id,
                'similarity': float(sim),
                'drug_name': metadata['drug_name'],
                'chain_type': metadata['chain_type'],
                'therapeutic_area': metadata.get('therapeutic_area', 'Unknown'),
                'metadata': metadata
            })
            
            if len(results) >= top_k:
                break
        
        search_time = time.time() - start_time
        return results
    
    def get_database_stats(self):
        chain_types = {}
        therapeutic_areas = {}
        
        for meta in self.metadata:
            chain_type = meta['chain_type']
            chain_types[chain_type] = chain_types.get(chain_type, 0) + 1
            
            area = meta.get('therapeutic_area', 'Unknown')
            therapeutic_areas[area] = therapeutic_areas.get(area, 0) + 1
        
        return {
            'total_proteins': len(self.protein_ids),
            'unique_drugs': len(set([meta['drug_name'] for meta in self.metadata])),
            'embedding_dimension': self.embeddings.shape[1],
            'chain_types': chain_types,
            'therapeutic_areas': therapeutic_areas
        }

# Initialize search engine
try:
    search_engine = ProteinSimilaritySearch()
    stats = search_engine.get_database_stats()
    
    print(f"\n📊 Database Statistics:")
    for key, value in stats.items():
        print(f"  • {key}: {value}")
        
    print(f"\n✅ Search engine initialized successfully!")
    print(f"🚀 Ready to run performance analysis!")
    
except Exception as e:
    print(f"❌ Error initializing search engine: {e}")
    print(f"📍 Current directory: {os.getcwd()}")
    print(f"📁 Files in current directory: {os.listdir('.')}")
    raise

# Quick test search
print(f"\n🧪 Quick Test Search:")
test_protein = search_engine.protein_ids[0]
test_results = search_engine.search_similar(query_protein_id=test_protein, top_k=3)

print(f"Query: {test_protein}")
for i, result in enumerate(test_results, 1):
    print(f"  {i}. {result['protein_id']} - {result['similarity']:.3f}")

print(f"\n🎉 Everything working! Ready for full analysis!")
```
