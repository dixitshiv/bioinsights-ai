```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/drugbank_biologics.csv")
```

```python
print("🧬 BioInsights AI - Data Overview")
print("="*50)
print(f"Total sequences: {len(df)}")
print(f"Unique drugs: {df['name'].nunique()}")
print(f"Average sequence length: {df['sequence_length'].mean():.0f}")
```

```python
print("\n📊 Complete Dataset:")
print(df[['name', 'brand_name', 'chain_type', 'sequence_length', 'approval_year']])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='chain_type', y='sequence_length')
plt.title('Protein Sequence Lengths by Chain Type')
plt.ylabel('Sequence Length (amino acids)')
plt.show()
```

```python
print("\n🎯 Therapeutic Indications:")
for drug in df['name'].unique():
    indication = df[df['name'] == drug]['indication'].iloc[0]
    print(f"• {drug}: {indication}")
```
